```python
# A Python program that computes a cumulative grade-point average (CGPA)
def calculate_gpa():

    print('Welcome to the CGPA calculator.')
    print('Please enter all your letter grades and their credits ')

    #This dictionary maps letter grades to their corresponding grade points.
    points = {'S': 10, 'A': 9, 'B': 8, 'C': 7, 'D': 6, 'E': 5, 'F': 0, 'I': None, 'W': None}
    
    grades = []
    credits = []

    #This loop continues to prompt the user for grades and credits.
    #If an entered grade is not in the dictionary or the credit is not a valid integer, an error message is printed and the input is ignored.

    while True:
        grade = input("Enter Grades ").strip().upper()
        if grade =='':
            break
        elif grade not in points:
            print("Unknown grade '{0}' being ignored".format(grade))
            continue
    
        credit = input("Enter credit for this course").strip()
        if not credit.isdigit():
            print(f"Invalid credit 'credit' being ignored")
            continue
        credit = int(credit)
        grades.append(grade)
        credits.append(credit)

    #If no valid grades are entered, the program exits. 
    #Otherwise, it calculates the total grade points and total credits, then computes and prints the TGPA.
    if not grades:
        print("Not valid grade entered. Exiting")
        return
    
    total_points = sum(points[grade] * credit for grade, credit in zip(grades, credits) if points[grade] is not None)
    total_credits = sum(credit for grade, credit in zip(grades, credits) if points[grade] is not None)
    
    if total_credits == 0:
        print('No valid grades entered. Exiting.')
        return
    
    tgpa = total_points / total_credits
    print(f'Your TGPA is {tgpa:.2f}')

```

```python
calculate_gpa()
```

```python
#Write a short Python function, minmax(data), that takes a sequence of one or more numbers, and returns the smallest and largest numbers, in the form of a tuple of length two. Do not use the built-in functions min or max in implementing your solution.
def minmax(data):
    if not data:
        raise ValueError("The data sequence must contain at least one element.")

    min_val = data[0]
    max_val = data[0]

    for num in data:
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num

    return (min_val, max_val)
```

```python
# Write a pseudo-code description of a function that reverses a list of n integers, so that the numbers are listed in the opposite order than they were before, and compare this method to an equivalent Python function for doing the same thing.
def reverse_list(data):
    start = 0
    end = len(data) - 1
    
    while start < end:
        # Swap elements at start and end
        data[start], data[end] = data[end], data[start]

        # Move the pointers
        start += 1
        end -= 1

    return data
```

```python
data = [ 4 , 5 , 7 , 9 , 6 ]
result = reverse_list(data)
print(result)
```

```python
# Write a short Python function that takes a sequence of integer values and determines if there is a distinct pair of numbers in the sequence whose product is odd.

def has_odd_product_pair(data):
    # Count the number of odd numbers in the sequence
    odd_count = 0
    
    for num in data:
        if num % 2 != 0:
            odd_count += 1
        # If we have found at least two odd numbers, return True
        if odd_count >= 2:
            return True
    
    # If we have less than two odd numbers, return False
    return False

# Example usage:
sequence = [2, 4, 6, 7, 9]
result = has_odd_product_pair(sequence)
print(result)  
```

```python
#Write a Python function that takes a sequence of numbers and determines if all the numbers are different from each other (that is, they are distinct)

def all_distinct(data):
    seen = set()
    for num in data:
        if num in seen:
            return False
        seen.add(num)
    return True


sequence1 = [1, 2, 3, 4, 5]
sequence2 = [1, 2, 2, 3, 4]

print(all_distinct(sequence1))
print(all_distinct(sequence2))
```

```python
import random

def my_shuffle(data):
    # Start from the last element and swap backwards
    for i in range(len(data) - 1, 0, -1):
        # Pick a random index j where 0 <= j <= i
        j = random.randint(0, i)
        # Swap data[i] with data[j]
        data[i], data[j] = data[j], data[i]

my_list = [1, 2, 3, 4, 5]
print("Original list:", my_list)
my_shuffle(my_list)
print("Shuffled list:", my_list)
```

```python
# Write a short Python function that counts the number of vowels in a given character string.

def count_vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

example_string = "Hello, World!"
vowel_count = count_vowels(example_string)
print(f"The number of vowels in '{example_string}' is {vowel_count}.")
```

```python
#Write a short Python function that takes a strings,representing a sentence, and returns a copy of the string with all punctuation removed. For exam- ple, if given the string "Let s try, Mike.", this function would return "Lets try Mike".

import string

def remove_punctuation(s):
    return ''.join([char for char in s if char not in string.punctuation])

example_string = "Let's try, Mike."
clean_string = remove_punctuation(example_string)
print(f"Original: {example_string}")
print(f"Without punctuation: {clean_string}")
```

```python
# Write a short Python program that takes two arrays a and b of length n storing int values, and returns the dot product of a and b. That is, it returns an array c of length n such that c[i] = a[i]·b[i], for i = 0,...,n−1.

def dot_product(a, b):
    # Check if both arrays have the same length
    if len(a) != len(b):
        raise ValueError("Both arrays must have the same length")
    
    # Compute the dot product
    c = [a[i] * b[i] for i in range(len(a))]
    return c

# Example usage:
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]

result = dot_product(a, b)
print(f"The dot product of {a} and {b} is {result}")
```

```python
# Write a Python program that can take a positive integer greater than 2 as input and write out the number of times one must repeatedly divide this number by 2 before getting a value less than 2.

def count_divisions(n):
    if n <= 2:
        raise ValueError("Input must be a positive integer greater than 2.")
    
    count = 0
    while n >= 2:
        n /= 2
        count += 1
    
    return count

try:
    num = int(input("Enter a positive integer greater than 2: "))
    divisions = count_divisions(num)
    print(f"Number of divisions needed: {divisions}")
except ValueError as e:
    print(e)
```

```python
# Write a Python program that can “make change.” Your program should take two numbers as input, one that is a monetary amount charged and the other that is a monetary amount given. It should then return the number of each kind of bill and coin to give back as change for the difference between the amount given and the amount charged. The values assigned to the bills and coins can be based on the monetary system of Indian Government. Try to design your program so that it returns as few bills and coins as possible.

def make_change(amount_charged, amount_given):
    if amount_given < amount_charged:
        print("Error: Amount given is less than amount charged.")
        return
    
    # Indian currency denominations in descending order
    denominations = [2000, 500, 200, 100, 50, 20, 10, 5, 2, 1]
    change_amount = amount_given - amount_charged
    
    print(f"Change to be given: {change_amount} INR")
    print("Breakdown of bills and coins:")
    
    for denom in denominations:
        if change_amount >= denom:
            num = change_amount // denom
            change_amount %= denom
            if num > 0:
                print(f"{num} x {denom} INR")

amount_charged = float(input("Enter the amount charged (INR): "))
amount_given = float(input("Enter the amount given (INR): "))

make_change(amount_charged, amount_given)
```

```python
# Write a Python program that can simulate a simple calculator, using the console as the exclusive input and output device. That is, each input to the calculator, be it a number, like 12.34 or 1034, or an operator, like + or =, can be done on a separate line. After each such input, you should output to the Python console what would be displayed on your calculator.

def calculator():
    result = 0.0
    operator = ''
    operand = 0.0
    display = '0.0'

    while True:
        print(display)
        user_input = input().strip()

        if user_input.isdigit() or (user_input.startswith('-') and user_input[1:].isdigit()):
            # Input is a number
            if display == '0.0':
                display = user_input
            else:
                display += user_input
        elif user_input in ['+', '-', '*', '/']:
            # Input is an operator
            operator = user_input
            operand = float(display)
            display += f' {operator} '
        elif user_input == '=':
            # Calculate result
            second_operand = float(display.split()[2])
            if operator == '+':
                result = operand + second_operand
            elif operator == '-':
                result = operand - second_operand
            elif operator == '*':
                result = operand * second_operand
            elif operator == '/':
                try:
                    result = operand / second_operand
                except ZeroDivisionError:
                    display = "Error: division by zero"
                    continue
            display = str(result)
        elif user_input.lower() == 'clear':
            # Clear the calculator
            result = 0.0
            operator = ''
            operand = 0.0
            display = '0.0'
        elif user_input.lower() == 'exit':
            # Exit the calculator
            break
        else:
            # Invalid input
            print("Invalid input. Please enter a number, operator (+, -, *, /), 'clear' to reset, or 'exit' to quit.")

if __name__ == "__main__":
    calculator()
```

```python
# Write a Python program that inputs a list of words, separated by white- space, and outputs how many times each word appears in the list. You need not worry about efficiency at this point, however, as this topic is something that will be addressed later in this book.

def count_word_occurrences(words):
    """Count occurrences of each word in the input list."""
    word_count = {}
    
    # Split the input string into word
    word_list = words.split()
    
    # Count occurrences of each word
    for word in word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    
    return word_count

def main():
    # Input a line of words separated by whitespace
    input_words = input("Enter a list of words separated by whitespace: ")
    
    # Count occurrences of each word
    word_count = count_word_occurrences(input_words)
    
    # Print the results
    print("\nWord Counts:")
    for word, count in word_count.items():
        print(f"{word}: {count}")

if __name__ == "__main__":
    main()
```

```python
#The birthday paradox says that the probability that two people in a room will have the same birthday is more than half, provided n, the number of people in the room, is more than 23. This property is not really a paradox, but many people find it surprising. Design a Python program that can test this paradox by a series of experiments on randomly generated birthdays, which test this paradox for n = 5,10,15,20,...,100.

import random

def generate_birthdays(num_people):
    """Generate random birthdays for num_people people."""
    birthdays = []
    for _ in range(num_people):
        # Generate a random birthday (day of year)
        birthday = random.randint(1, 365)
        birthdays.append(birthday)
    return birthdays

def has_duplicates(birthdays):
    """Check if there are any duplicate birthdays in the list."""
    # Using a set to track seen birthdays
    seen = set()
    for day in birthdays:
        if day in seen:
            return True
        seen.add(day)
    return False

def birthday_paradox_experiment(num_people, num_trials):
    """Run the birthday paradox experiment for given number of people."""
    num_success = 0
    for _ in range(num_trials):
        birthdays = generate_birthdays(num_people)
        if has_duplicates(birthdays):
            num_success += 1
    return num_success / num_trials

def main():
    # Define the range of number of people to test
    min_people = 5
    max_people = 100
    step = 5
    
    # Number of trials for each experiment
    num_trials = 10000
    
    print("Testing the Birthday Paradox...")
    print(f"Number of trials per experiment: {num_trials}\n")
    print("  n    Probability of shared birthday")
    print("-------------------------------------")
    
    for n in range(min_people, max_people + 1, step):
        probability = birthday_paradox_experiment(n, num_trials)
        print(f"  {n:2}    {probability:.4f}")
    
if __name__ == "__main__":
    main()
```

```python
#A common punishment for school children is to write out a sentence multiple times. Write a Python stand-alone program that will write out the following sentence one hundred times: “I will never spam my friends again.” Your program should number each of the sentences and it should make eight different random looking typos.


import random

def generate_typos(sentence):
    """Generate random-looking typos for a sentence."""
    # Split the sentence into words
    words = sentence.split()
    num_words = len(words)
    
    # If there are less than 8 words, use all words for typo introduction
    if num_words < 8:
        typo_indices = list(range(num_words))  # Use all words
    else:
        typo_indices = random.sample(range(num_words), 8)  # Sample 8 random indices
    
    # Introduce typos at selected indices
    for idx in typo_indices:
        word = words[idx]
        # Introduce a random typo (replace a character randomly)
        if len(word) > 1:
            char_idx = random.randint(0, len(word)-1)
            new_char = chr(random.randint(97, 122))  # Random lowercase ASCII character
            words[idx] = word[:char_idx] + new_char + word[char_idx+1:]
    
    # Join the words back into a sentence
    return ' '.join(words)

def main():
    sentence = "I will never spam my friends again."
    
    for i in range(1, 101):
        # Generate typos
        sentence_with_typos = generate_typos(sentence)
        
        # Print numbered sentence
        print(f"{i}. {sentence_with_typos}")

if __name__ == "__main__":
    main()
```
